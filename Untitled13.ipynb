```python

spark_passes = spark.createDataFrame(top_passes)

spark_dribles = spark.createDataFrame(top_dribles)

spark_chutes = spark.createDataFrame(top_chute_gol)

spark_faltas = spark.createDataFrame(top_jogador_faltas)
```

```python
def limpar(df, coluna_valor):
    return (
        df.dropDuplicates()                # Remove duplicados
          .dropna()                         # Remove nulos
          .filter(col(coluna_valor) > 0)    # Remove valores inválidos
          .withColumn(coluna_valor, col(coluna_valor).cast("int"))  # Garante tipo correto
          .withColumn("jogador", trim(col("jogador")))             # Padroniza nome
    )
```

```python

spark_passes = limpar(spark_passes, "passes")
spark_dribles = limpar(spark_dribles, "dribles")
spark_chutes = limpar(spark_chutes, "chutes")
spark_faltas = limpar(spark_faltas, "faltas")
```

```python
print("Spark Passes (Cleaned):")
spark_passes.printSchema()
spark_passes.show()

print("Spark Dribles (Cleaned):")
spark_dribles.printSchema()
spark_dribles.show()

print("Spark Chutes (Cleaned):")
spark_chutes.printSchema()
spark_chutes.show()

print("Spark Faltas (Cleaned):")
spark_faltas.printSchema()
spark_faltas.show()
```

```python
print("=== DataFrame de Passes (Spark) ===")
spark_passes.show()
```

```python
top10_passes_spark = spark_passes.orderBy(col("passes").desc()).limit(10)
print("=== TOP 10 PASSES (Spark) ===")
top10_passes_spark.show()
```
